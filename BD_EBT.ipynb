```python
# Conectando o Colab ao Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Importando o pandas
import pandas as pd
```

```python
selecao = ['UF', 'Código IBGE', 'Ranking', 'Nota Avaliação']
df1 = pd.read_csv('/content/drive/MyDrive/Template Dados/input/EBT_1_edicao.csv', sep=';', dtype = str)
df1 = df1[selecao]
df1['ano'] = '2020'
df2 = pd.read_csv('/content/drive/MyDrive/Template Dados/input/EBT_2_edicao.csv', sep=';', dtype = str)
df2 = df2[selecao]
df2['ano'] = '2021'
df = df1.append(df2, ignore_index=True)
df.rename(columns={'UF' : 'sigla_uf', 'Código IBGE':'id_municipio', 'Ranking':'ranking', 'Nota Avaliação':'nota'}, inplace=True)


ufs = {'Acre':'AC', 'Alagoas':'AL', 'Amazonas':'AM', 'Amapá':'AP', 'Bahia':'BA', 'Ceará':'CE', 'Distrito Federal':'DF', 'Espírito Santo':'ES', 'Goiás':'GO', 
       'Maranhão':'MA', 'Minas Gerais':'MG', 'Mato Grosso do Sul':'MS', 'Mato Grosso':'MT', 'Pará':'PA', 'Paraíba':'PB', 'Pernambuco':'PE', 'Piauí':'PI', 
       'Paraná':'PR', 'Rio de Janeiro':'RJ', 'Rio Grande do Norte':'RN', 'Rondônia':'RO', 'Roraima':'RR', 'Rio Grande do Sul':'RS', 
       'Santa Catarina':'SC', 'Sergipe':'SE', 'São Paulo':'SP', 'Tocantins':'TO'} # Renomeando a sigla_uf

df['sigla_uf'] = df['sigla_uf'].map(ufs) # Criar uma estrutura de repetição, a qual todos os ufs são integrados em sigla_uf
df = df[['ano', 'sigla_uf', 'id_municipio', 'ranking', 'nota']] # Organizando as colunas (Tal, qual, arquitetura dos dados)


df['id_municipio'] = df['id_municipio'].fillna('0') # Substitui os valores nulos em id_municipio por 0
uf = df.query('id_municipio == "0"')  # Usando o Query para encontrar id_municipio == 0


uf.drop('id_municipio', axis=1, inplace=True) # excluir axis = 1 (coluna)
uf.to_csv('/content/drive/MyDrive/Template Dados/output/uf.csv', index=False, na_rep='') # # Subir o arquivo csv para o drive na pasta input
municipio = df.query('id_municipio != "0"') # Criando uma variável 
municipio.to_csv('/content/drive/MyDrive/Template Dados/output/municipio.csv', index=False, na_rep='') # Subir o arquivo csv para o drive na pasta output
```
